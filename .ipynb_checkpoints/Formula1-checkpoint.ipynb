```python
races_1_df = pd.read_csv("data/races.csv")
races_1_df
```

```python
#sorting dataset in chronological 'date' order
races_1_df = races_1_df.sort_values('date')
races_1_df
```

This file contains data for all the races that have taken place and will take place in the future. Removing the races that are yet to take place. The last completed race was the new addition to the F1 calendar, the Saudi Arabian GP held on 5th Decemeber 2021 at the Jeddah Corniche Circuit.

```python
#deleting the races that are still yet to take place
races_df = races_1_df[races_1_df.date <= '2021-12-12']

races_completed = len(races_df)
print('The number of chequered flags waved in F1 so far (until 2021 Abu Dhabi GP) is {}.'.format(races_completed))
```

```python

idx = races_df.groupby(['year'])['date'].transform(max) == races_df['date'] # Finds if the race index is last of the season
season_finale = races_df[idx].rename(columns = {'round' : 'tot_races'}) 
season_finale = season_finale[season_finale.year <= 2021] # A new data frame with season-ending races only, excluding the 2020 season

# Plotting
plt.figure(figsize=(15, 5))

plt.plot(season_finale.year, season_finale.tot_races, 's-b')
plt.xlabel('Year')
plt.ylabel('Number of races')
plt.title("Championship races per seasons");
```

## Circuits

##### Circuits around the world that have hosted Formula 1 races

```python
circuits_1_df = pd.read_csv('data/circuits.csv')
circuits_1_df.head()
```

Some of these circuits haven't held any race at the moment, for example the Vietnam GP held at the Hanoi Street Circuit but was cancelled due to the COVID-19 pandemic and various political issues surrounding the circuit.

```python
host_circuits = races_df.drop(['date','time','url','round'], axis = 1).rename(columns = {'name': 'gp_name'}).merge(
    circuits_1_df.drop(['lat','lng','alt','url'], 1).rename(columns = {'name':'cir_name'}), how = 'left')
```

##### How many circuits have actually hosted a proper race 

```python
# merging the data frame with the one containing data of all the races.
tot_host_circuit = host_circuits['circuitId'].nunique()
print('{} different circuts around the world have hosted a Formula 1 Grand Prix on their tarmac'.format(tot_host_circuit))
```

So far 76 circuits have played venues for F1 races. Jeddah Corniche Circuit is the latest addition to this list after it hosted the Saudi Arabian Grand Prix. 

###### Top 10 circuits to host a GP

```python
#Sorting the top 10 races based on occurance
top_host = host_circuits[['cir_name']].cir_name.value_counts().reset_index().rename(
        columns = {'index' : 'name', 'cir_name' : 'counts'}).head(10)

#Plotting
plt.figure(figsize = (12, 6))

sns.barplot(top_host.counts, top_host.name)
plt.title('Top 10 Formula 1 host circuits')
plt.xlabel('Number of Races hosted')
plt.ylabel('');
```

With 71 races, Autodromo Nazionale Monza has hosted more GPs than any other circuit. The 'Temple of Speed' racetrack has been a part of every Formula1 season, except in the 1980 season. In the second spot, we have the legendary Circuit de Monaco laid out on the streets of Monte Carlo. Silverstone and Circuit de Spa-Francorchamps are other iconic circuits that can boast more than 50 races. All of these historic tacks can still be found on present Formula 1 calendars.

## Constructors

Being a technology-driven sport, constructors have a massive role in F1. A driver can only be as good as the car underneath him. F1 also runs a championship for the constructors along with the drivers' championship. Let's look at all the players that have been in this business.

```python
constructors_1 = pd.read_csv('data/constructors.csv')
constructors_1
```

Here we have an 211 official entrants till 2021, technically this is the list of all the entrants/teams and not constructors. 

```python
results_1_df = pd.read_csv('data/results.csv')
results_1_df
```

```python
#Creating a function to show driver data

def podium_counter(pos):
    if pos in ['1','2','3']:
        return True
    else:
        return False

def win_counter(pos):
    if pos == '1':
        return True
    else:
        return False
    
def pole_counter(grid):
    if grid == 1:
        return True
    else:
        return False
```

```python
results_copy = results_1_df[['driverId', 'raceId', 'grid', 'position', 'points', 'laps', 'rank']].copy().rename(
    columns = {'rank' : 'fastlap'})

#Adding podium, win, pole and fastest lap
results_copy['podium'] = results_copy.position.apply(podium_counter)
results_copy['win'] = results_copy.position.apply(win_counter)
results_copy['pole'] = results_copy.grid.apply(pole_counter)
results_copy['fastestLap'] = results_copy.fastlap.apply(win_counter)

#Grouping with respect to drivers and applying rules to different columns
driver_stats_1 = results_copy.drop(columns = ['position', 'grid', 'fastlap']).groupby('driverId')
func_dic = {'raceId' : 'count','points' : 'sum','laps' : 'sum','podium' : 'sum','pole' : 'sum','win' : 'sum', 'fastestLap' : 'sum'}
driver_stats_1 = driver_stats_1.aggregate(func_dic).reset_index().rename(columns = {'raceId':'races'})
```

```python
#Getting Drivers Championship Data

driver_st_1 = pd.read_csv('data/driver_standings.csv')
driver_st_season_end = season_finale[['raceId', 'year' , 'tot_races']].merge(driver_st_1, on = 'raceId')
driver_champ = driver_st_season_end[driver_st_season_end['position'] == 1]
driver_champ_tot = driver_champ[['driverId', 'position']].groupby('driverId').sum().reset_index().merge(
drivers_1[['forename', 'surname', 'driverId']]).rename(columns = {'position':'titles'}).sort_values('titles' , ascending = False)

#Merging everything into one dataframe
driver_stats = driver_stats_1.merge(driver_champ_tot[['driverId' , 'titles']], how = 'left').fillna(0)
driver_stats = drivers_1[['driverId', 'forename', 'surname', 'nationality']].merge(driver_stats, on = 'driverId')
driver_stats
```

```python
#function to display driver details

def get_driver_stats():
    f_n,s_n = input("Enter Driver Name :").split()
    df = driver_stats.loc[(driver_stats['forename'] == f_n) & (driver_stats['surname'] == s_n)].squeeze()
    return print(
'''Name: {} {}
    Nationality: {}
    Stats
    Races Entered:\t{}
    Drivers' Titles:\t{:.0f}
    Race Wins:\t \t{}
    Pole Positions:\t \t{}
    Podiums:\t \t{}
    Total Points:\t \t{}
    Fastest Laps:\t \t{}
    Laps Raced:\t \t{}'''.format(df.forename, df.surname, df.nationality, df.races, df.titles, df.win, df.pole, 
                                  df.podium, df.points, df.fastestLap, df.laps))
```

Statistically, Hamilton is the greatest driver of all time, having equaled Michael Schumacher's 7 World Championship Records despite competing in fewer races. Hamilton holds numerous records, including the most wins, pole positions, podium finishes, points scored, championships won, laps led, and being the only driver to win a race in each season.

Each driver hails from a distinct era of Formula One. The modern era is represented by Lewis Hamilton and Sebastian Vettel, while the 1990s and 2000s were dominated by Michael Schumacher. Schumacher's F1 numbers are incredible. He has seven world titles, 91 victories, 155 podiums, and 68 pole positions to his credit. Lewis Hamilton, on the other hand, appears to be on his way to breaking all records.

Looking at one of Lewis Hamilton's most remarkable achievements, he has won at least one Grand Prix every year since his debut in 2007. Let's take a look at his victories over the years.

```python
driver_year_end = season_finale[['raceId', 'year']].merge(driver_st_1)
lewis_year_end = driver_year_end[driver_year_end['driverId'] == 1]

#Plotting

plt.figure(figsize = (16,6))
plt.bar(lewis_year_end.year, lewis_year_end.wins)
plt.xlabel('Year')
plt.ylabel('')
plt.title("Lewis Hamilton's year-round wins");
```

Lewis came within a whisker of winning the world title in his rookie season in 2007, losing by a single point, before winning his first title in 2008. From 2007 to 2012, his win total with McLaren is shown. He joined Mercedes in 2013, just before the turbo-hybrid era began. With only one win in his first year with Mercedes, it was probably the low point of his career. But he didn't look back after that, winning nearly half of all races and a whopping six World Championships in the following seasons.
